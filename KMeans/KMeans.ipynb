### KMeans

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import random
```

```python
dataset = pd.read_csv('./Iris.csv')
name = list(set(dataset['class']))
dataset.drop(['sepallength','sepalwidth','class'],axis='columns',inplace=True)
scalar = MinMaxScaler()
scalar.fit(dataset[['petallength']])
dataset['petallength'] = scalar.transform(dataset[['petallength']])
scalar.fit(dataset[['petalwidth']])
dataset['petalwidth'] = scalar.transform(dataset[['petalwidth']])
model = KMeans(n_clusters = 3)
predicted = model.fit_predict(dataset)
```

```python
dataset['cluster'] = predicted
dataset1 = dataset[dataset['cluster']==0]
dataset2 = dataset[dataset['cluster']==1]
dataset3 = dataset[dataset['cluster']==2]
plt.scatter(dataset1['petallength'],dataset1['petalwidth'],color='orange',marker='d',label=name[0])
plt.scatter(dataset2['petallength'],dataset2['petalwidth'],color='blue',marker='*',label=name[1])
plt.scatter(dataset3['petallength'],dataset3['petalwidth'],color='green',label=name[2])
plt.scatter(model.cluster_centers_[0][0],model.cluster_centers_[0][1],color='red',marker = '+',label='center')
plt.scatter(model.cluster_centers_[1][0],model.cluster_centers_[1][1],color='red',marker = '+')
plt.scatter(model.cluster_centers_[2][0],model.cluster_centers_[2][1],color='red',marker = '+')
plt.legend()
plt.show()
```
